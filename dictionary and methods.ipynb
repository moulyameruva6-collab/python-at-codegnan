```python
#Dictionary can't allow list as key
details={'name':'Moulya','age':20,'course':'python','Marks':[80,90,86],'Age':50}
print(details)
```

```python
#Dictionary can't allow list as key
#If dictionary contains multiple items with same key,In that case it updates first item key value with last defined item value
details={'name':'Moulya','age':20,'course':'python','Marks':[80,90,86],'age':50}
print(details)
```

```python
#access values by using keys
details={'name':'Moulya','age':20,'course':'python','Marks':[80,90,86],'age':50}
print(details['course'])
print(details['age'])
```

```python
#access values by using keys:
details={'name':'Moulya','age':20,'course':'python','Marks':[80,90,86],'age':50}
print(details['course'])
print(details['age'])
#Updating the dictionary value:
details['age']+=1
details['course']='java'
print(details)
```

```python
#access values by using keys:
details={'name':'Moulya','age':20,'course':'python','Marks':[80,90,86],'age':50}
print(details['course'])
print(details['age'])
#Updating the dictionary value:
details['age']+=1
details['course']='java'#It updating the values for 'course'key
print(details)
#Item assignment:
details['location']="India"#It adds new items to dictionary
print(details)
```

```python
#define empty Dictionary:
d1={}
d2=dict()
print(type(d1),type(d2))
```

```python
###DICTIONARY METHODS:
-Gets value
 -get(keys,default_val):It returns the value if keys exists in dictionary otherwise
default_val
-Update values and add items to dictionary 
 - update (new_dictionary):dictionary updates with New_dictionary items 
-Remove items from dictionary
 -pop(key):it returns value and remove item if key is exists in dictionary
otherwise 'KeyError'
  -popitem():it returns tuple of key value pair and removes the last item from dictionary
-Get keys,value and items from dictionary
  -keys():it returns list of dictionary keys
  -values():it returns list of dictionary values
  -items():it returns list of tuple key-value pairs
```
