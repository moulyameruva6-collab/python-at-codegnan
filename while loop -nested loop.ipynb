```python
N = int(input())
l=[]
for c in range(N):
    op=input().split()
    if op[0] == "insert":
        l.insert(int(op[1],int(op[2])))
    elif op[0] == "remove":
        l.remove(int(op[1])) 
    elif op[0] == "append":
        l.append(int(op[1]))
    elif op[0] == "pop":
        l.pop()
    elif op[0] == "reverse":
        l.reverse()
    elif op[0] == "print":
        print()
    elif op[0] =="sort":
        l.sort()
           
```

```python
#nested loop:
n=3
num = 1
for r in range(1,n+1):
    for c in range(1,n+1):
        print(num,end=" ")
        num += 1 
    print()
```

```python
# even and odd numbers in pattern format:
even = 2
odd = 1
for r in range(3):
    for c in range(3):
        if r % 2 == 0:
            print(even,end=" ")
            even += 2
        else:
            print(odd,end=" ")
            odd += 2
    print()
        
```
