```python
## 2. Enclosing Scope::
def outer(x): #Enclosing Scope
    x = 10
    def inner():
        x = 20 # Local Scope
        z = x+x
        print(f"This is Local Scope:{z}")
    inner()
    y = x+x
    print(f"This is enclosing scope:{y}")
n = 5
outer (n)
```

```python
## 3. Global Scope::
x = 10 #Global scope
def outer():
    y = 20 #Enclosing scope
    def inner():
        z = 30 # Local scope
        res = z+y+x
        print(f"this is local scope:{res}")
    y =y+x
    inner()
    print(f"the is enclosing scope:{y}")
outer()
```

```python
# Global scope:
# trying to update the Global scope variable inside the local scope:
x = 10 # Global scope
def outer():
    #x = 20
    x = x+1
    print("this is local scope val:",x)
outer()
print("this is global scope val:",x)
```

```python
# update the Global scope variable inside the local scope by using 'global' keyword:
x = 10 # Global scope
def outer():
    global x
    x = x+1
    print("this is local scope val:",x)
outer()
print("this is global scope val:",x)
```
