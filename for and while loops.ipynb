```python
#count even and odd numbers in given list of numbers:
arr=list(map(int,input().split()))
even=odd=0
for num in arr:
    if num % 2 == 0:
        even  += 1
    else:
        odd += 1
print(even,odd)
```

```python
##seperate the even and odd numbers in given list
arr=list(map(int,input().split()))
even = []
odd = []
for num in arr:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)
print(even,odd)
```
