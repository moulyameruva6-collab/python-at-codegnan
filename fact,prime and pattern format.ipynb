```python
#factorial and prime number:
num = int(input())
factors=[]
for fact in range(1,num+1):
    if num % fact == 0:
        factors.append(fact)
        print(fact)
        if len(factors) == 2:
            print(f'{num} is prime')
else:
            print(f'{num} is not prime')
```

```python
# find reverse of a string:
string = input()
rev = ""
for char in string:
    rev=char+rev
    print(rev)
```

```python
# find reverse and palindrome of a string:
string = input()
rev = ""
for char in string:
    rev=char+rev
print(rev)
if rev == string:
    print(f"{string} is palindrome")
else:
    print(f"{string} is not palindrome")
```

```python
rows = 9
cols = 6
ch = 65      # Start from A
forward = True
for i in range(rows):
    for j in range(cols):
        print(chr(ch), end=" ")
        if forward:
            ch += 1
            if ch > 90:      # After Z
                ch = 90       # Start reverse from Z
                forward = False
        else:
            ch -= 1
            if ch < 65:       # After A
                ch = 65       # Start forward again
                forward = True
    print()
```
