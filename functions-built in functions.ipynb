```python
## Built in functions: the function defination is provided by python ,those functions called built in functions.
```

```python
# 1. abs():it returns the absulute value
print(abs(20-10))
print(abs(10-20))
```

```python
# 2. round(val,pos): this function rounds the float values particular positions
print(round(2.2589,2))
print( round(2.2357,2))
print(round(2.2539,2))
print(round(2.2539))
```

```python
# 3. ord(char): it returns the AscII valuesof characters
#char(ascii_val): it return the character for specified ASCII value
print(ord("A"))
print(chr(65))
```

```python
# 4. sorted(iterable,reverse=false,key=function):it returns new sorted iterable
# sort the fruit based on the length of fruit name
fruits = ['Apple','Banana','Kiwi','Mango','Water Melon']
res = sorted(fruits)
print(res)
print(sorted(fruits,reverse = True))
print(sorted(fruits,reverse = True,key = len))# it sorts the list based on length of each variable
```

```python
# 5. map(function , iterable):
# find the square of a all list of integers using map functions
def square(x):
    return x**2
arr = [1,2,3,4,5]
res = list(map(square,arr))
print(res)
```

```python
# find the even numbers from list of numbers
def check_even(x):
    if x % 2 == 0:
        return x
lst = [1,2,3,4,5,6]
res = list(map(check_even,lst))
print(res)
```

```python
# 6. filter(function , iterable):it filters the iterable,if the function return non-none values,those values will added to result.
# find the even numbers from list of numbers
def check_even(x):
    if x % 2 == 0:
        return x
lst = [1,2,3,4,5,6]
res = list(filter(check_even,lst))
print(res)
```

```python
# find the 3's multipliers from list of integers
arr =[1,24,54,22,44,51,33,15]
def three_multipliers(x):
    if x % 3 == 0:
        return 1
res = list(filter(three_multipliers,arr))
print(res)
```

```python
# 7. enumerate(iterable): this function return the list of tuples of index and items for iterable items
fruits = ['mango' , 'apple' , 'banana' , 'kiwi' , 'water melon']
print(enumerate(fruits))
print(list(enumerate(fruits)))
```

```python
# 7. enumerate(iterable): this function return the list of tuples of index and items for iterable items
fruits = ['mango' , 'apple' , 'banana' , 'kiwi' , 'water melon']
print(enumerate(fruits))
print(list(enumerate(fruits)))
for ind, val in enumerate(fruits):
    print(f"{ind}-{val} and length is {len(val)}")
```

```python
# 8. tuple unpacking:
t =(1,2)
a , b = t
print(a,b)
```

```python
# 9. zip(iterable1,iterable2,....):it returns the all iterable in asingle iterable in tuples
rool_no = [101,102,103,104,105,106]
names = ['moulya','swathi','divya','anjali','swetha','lalitha']
percentage = [98,89,76,84,96,87]
print(list(zip(rool_no,names,percentage)))
```

```python
percentage = [34,35,64,85,76]
lst=[]
for num in percentage:
    if num < 35:
        lst.append(False)
    else:
        lst.append(True)
print(lst)
if all(lst):
    print("all students pass")
else:
    print("all students are not passed")
```
