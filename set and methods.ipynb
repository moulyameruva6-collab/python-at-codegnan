###set methods:
- Adding elements to set
  - add(ele):it adds element to set element not exists in set,otherwise nothing
  - update(iterable):it adds all iterable elements in sets
- Remove elements from set
  - remove(ele):it removes the element from set if element exists in set ,otherwise 'KeyError'
 - discard(ele):it removes the element from set if element exists in set,otherwise nothing
 - pop():it remove and returns elements from set 
- SET OPERATIONS
  - union(anthor_sets):it returns the all common elements from all sets
  - intersection(anthor_sets):it returns the common elements from all sets
  - difference(anthor_sets):it returns the set elements those are not lies in anthor_sets
  - symmetric_difference(anthor_sets):it returns the non-common elements from all sets


```python
s={1,2,3,-5}
s.add(4)
print(s)
s.add(1)# 1 doesn't add to set because set contains 1
print(s)
s.update([10,20,30])
print(s)
print(len(s))
s.remove(1)
print(s)
#s.remove(1)#KeyError
s.discard(1)
print(s)
```

```python
s1={3,5,2,7}
s2={2,4,5,6,3}
s3={2,4,3}
print(s1|(s2))#|--union
print(s1|(s2|s3))
print(s2&(s3))#&--intersection
print(s2&(s1))
print(s1-(s2-s3))#- difference
print(s1-(s2))
print(s2-(s3))
print(s1^(s2))#^-symmetric difference
```
