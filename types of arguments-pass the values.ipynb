```python
# Types of Arguments:
#1.Positional Argument: the arguments are assign to perameters as per the positional order
def sub(x:int,y:int)->int:
    c = x - y
    return c 
#function calling block
a , b = map(int,input().split())
res = sub(a,b)
print(res)
print(sub(b,a))
    
```

```python
# 2. Keyword Arguments:here arguments passed along with the parameters:
def sub(x:int,y:int)->int:
    c = x - y
    return c 
#function calling block
a , b =10 , 20
print(sub(x=a,y=b))
print(sub(y=b,x=a))
    
```

```python
#sum of two numbers:
def add(x:int,y:int)->int:
    c = x + y
    return c 
#function calling block
a , b =10 , 20
res = add(a , b)
print(res)
print(add(b,a))
    
```

```python
#sum of three numbers:
def add(x:int,y:int,z:int)->int:
    c = x + y+z
    return c 
#function calling block
a , b  ,c =10 , 20 , 30
res = add(a , b , c)
print(res)
```

```python
# multiply the numbers::
def mul(a,b,c=1,d=2):
    res = a*b*c*d
    return res
print(mul(10,20))
print(mul(10,20,5))
print(mul(10,20,2,1))

    
```

```python
# 3.Default Arguments::the parameters defined with same default values , and  it assign the arguments value f you pass the values for  default parameters otherwise,it will assign default values::
def add(a,b,c=0,d=0):
    res = a+b+c+d
    return res
print(add(10,20))
print(add(10,20,5))
print(add(10,20,2,1))
print(add(10,30))
    
```

```python
# 4 .Variable Length Of Positional Arguments:
def add(a,b=0,*args):
    print(type(args))
    print(args)
print(add(1,2,3,4,5,6,7,8,9))
```

```python
# 4 .Variable Length Of Positional Arguments:
def add(a ,b=0,*args):
    print(type(args))
    print(args)
    s = 0
    for num in args:
        s += num
    return s + a + b
print(add(1))
#print(add(1 , 2 ,b=3))
print(add(1,2,3,5,5))
print(add(1,2,3,4,5,6,7,8,9))
```

```python
#pass by value::
def fun(x):
    print(x)
    print(f" Before update Address of x:{id(x)}")
    x = x + 1
    print(x)
    print(f" After update the x value, Address of x:{id(x)}")
a = 5
print(f" Before  function execution Address of a:{id(a)} and value: {a}")
print(fun(a))
print(f" After  function execution Address of a:{id(a)} and value: {a}")
```
