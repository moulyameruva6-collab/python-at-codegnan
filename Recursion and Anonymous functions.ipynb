```python
# 1. find the factorial of a number n:
def factorial(n):
    if n == 1 or n == 0:
        return 1
    return n*factorial(n-1)
factorial(4)
```

```python
# 2. find the function of a number n:
def fabinoci(n):
    if n == 1 or n == 0:
        return 1
    return fabinoci(n-1)+fabinoci(n-2)
fabinoci(4)
```

```python
#3. sum of 'n' natural numbers using recursions:
def sum_of_natural(n):
    if n == 1:
        return 1
    return n+ sum_of_natural(n-1)
sum_of_natural(5)
```

```python
#4. binary number for decimal number using recursion:
def binary(n):
    if n == 1:
        return "1"
    if n == 0:
        return "0"
    return binary(n//2)+str(n%2)
binary(4)
```

## Anonymous functions:: the function don't have any name and this functions implement by using 'lambda' keyword.
## syntax: lambda arguments : expression

```python
# 1. find the square of the number using lambda functions::
square = lambda x : x**2
print(square(5))
```

```python
# 2. sum of two numbers using lambda::
add = lambda x,y : x+y
print(add(13,23))
```

```python
# 3. find the square of all list of integer using lambda and map()::
arr = [1,2,3,4,5,6]
res = list(map(lambda x: x**2,arr))
print(res)
```

```python
# 4. find even numbers in list using lambda and filter()::
arr = [1,2,3,4,5,6]
res = list(filter(lambda x : x%2==0,arr))
print(res)
```

```python
#5. find the whether the number is even or odd using lambda::
n = int(input())
even_or_odd = lambda x: f"{x}-odd" if x%2 else f"{x}-even"
print(even_or_odd(n))
```
