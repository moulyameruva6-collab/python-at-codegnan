```python
#find afunctiomn is prime or not a prime number:
def prime_check(n:int)->bool:
    if n<2:
        return False
    for i in range(2,n//2+1):
        if n%i==0:
            return False
    return True
num = int(input())
print(prime_check(num))
```

```python
#find afunctiomn is prime or not a prime number:
def prime_check(n:int)->bool:
    for i in range(2,n//2+1):
        if n%i==0:
            return False
    return True
for j in range(1,100):
    if prime_check(j):
        print(j,end=" ")
```
