```python
##remove targets elements from list of integer: using for loop::
arr = list(map(int,input().split()))
target = int(input())
res=[]
for num in arr:
    if num != target:
        res.append(num)
print(res)
```

```python
#using for loop::index method::
arr = list(map(int,input().split()))
target = int(input())
res=[]
n=len(arr)
for i in range(n):
    if arr[i] !=target:
        res.append(arr[i])
print(res)
```

```python
## while loop:
arr = list(map(int,input().split()))
target = int(input())
res=[]
n = len(arr)
i=0 # initialization
while i < n:
    if arr[i]!= target:
        res.append(arr[i])
print(res)
    
```
